```python
import pandas as pd
```

```python
PATH_DATA = '../analisis_food_delivery/'
df_trx = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv')
df_trx.head()
```

```python
resto_terjangkau = df_trx[(df_trx['food_price']<=50000)]
resto_terjangkau
```

```python
series_resto = resto_terjangkau.groupby(['resto_id']).size()
series_rating = resto_terjangkau.groupby(['resto_id','rating']).size()
df_resto_terjangkau = series_rating.unstack('rating')
df_resto_terjangkau['order_count'] = series_resto
df_resto_terjangkau
```

```python
series_price = resto_terjangkau.groupby(['resto_id']).food_price.sum()
df_resto_terjangkau['total_food_price'] = series_price
average_cost = df_resto_terjangkau['total_food_price']/df_resto_terjangkau['order_count']
df_resto_terjangkau['average_cost'] = average_cost
df_resto_terjangkau
```
